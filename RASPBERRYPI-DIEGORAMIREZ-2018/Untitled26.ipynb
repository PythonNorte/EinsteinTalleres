```python
%matplotlib notebook
"""
============================================
Silla de montar o parabolide hiperbolico
DR. GONZALO DURAN INENCO
3er Sprint de Python Cientifico INENCO UNSa. Julio 2018.-
============================================
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
#Z = np.sqrt(X**2 + Y**2)
Z = np.sin(X)+np.sin(Y) #grafica la funcion de la silla

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.viridis)

plt.show()
```
